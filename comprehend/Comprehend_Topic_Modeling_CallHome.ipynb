```python
from nltk.corpus import stopwords
stops = stopwords.words('english')
```

```python
stops.extend(['uh','thing','hm','mhm','yes','yep','oh','okay','no','um'])
```

```python
s3 = boto3.resource('s3')
```

```python
# get all files in s3 bucket
input_bucket = 'callhome.audio.dataset'
save_bucket = 'call.home.transcript.clean'
```

```python
def process_csv(obj,output_file,stop):
    # get object from s3
    a = pd.read_csv('s3://'+obj.bucket_name+'/'+obj.key)
    b = a.dropna(subset=['Text'])['Text'].values
    b = ' '.join(b).lower()
    
    # remove punct
    b = b.translate(str.maketrans('', '', string.punctuation))
    tokens_without_sw = [word for word in b.split() if not word in stops]
    s3.Object(save_bucket, obj.key.split('.')[0]+'.txt').put(Body=' '.join(tokens_without_sw))
```

```python
# get the file object
bucket = s3.Bucket(input_bucket)
for obj in bucket.objects.all():
    if obj.key.split('/')[0].split('.')[-1] == 'csv':
        
        print(obj)
        process_csv(obj,save_bucket,stops)
```
